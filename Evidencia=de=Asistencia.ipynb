```python
import numpy as np
import matplotlib.pyplot as plt

def mi_funcion_sen(vmax, dc, ff, ph, nn, fs):
    tt=np.arange(0,nn,1)*(1/fs)
    xx=dc + vmax*np.sin(2*np.pi*ff*tt + ph)
    return tt,xx 

#Sin Ruido

tt, xx = mi_funcion_sen(vmax = 1, dc = 0, ff = 1, ph=0, nn=100, fs = 100)

plt.plot(tt, xx)
plt.title("Funcion Senoidal")
plt.xlabel("Tiempo [t]")
plt.ylabel("Voltaje [V]")
plt.grid(ls="--")
plt.show()

# Con Ruido

nn=100

ruido=np.random.normal(0,1,nn)

tt1, xx1 = mi_funcion_sen(vmax = 1, dc = 0, ff = 1, ph=0, nn=100, fs = 100)

y=xx1 + ruido*0.1 #Lo multiplico por 0.1 para que no halla tanto ruido
plt.figure()
plt.plot(tt1, y)
plt.title("Funcion Senoidal")
plt.xlabel("Tiempo [t]")
plt.ylabel("Voltaje [V]")
plt.grid(ls="--")
plt.show()


     
snr=20

pr=1/(10**(snr/10)) #despeje 

u_n= np.random.normal(0,np.sqrt(2),nn)
#xxn= xx+u_n
#varianza=np.var(xx1)
#print(pr)
#print(varianza)
#varianza2=np.var(y)
#print(varianza2)
```
